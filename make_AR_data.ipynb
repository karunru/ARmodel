```julia
using Gadfly
```

```julia
M = 3
```

```julia
a = [0.5,-0.02,-0.4]
```

```julia
N = 1000
```

```julia
using Distributions
```

```julia
d = Normal(0,1)
```

```julia
srand(1)
```

```julia
x = zeros(N)
```

```julia
for i in 1:M
    x[i] = rand(d,1)[1]
end
```

```julia
x
```

```julia
for i in M+1:N
    for j in 1:M
        x[i] += a[j]*x[i-j]
    end
    x[i] += rand(d,1)[1]
end
```

```julia
x
```

```julia
t = 1:N
```

```julia
plot(x=t,y=x,Geom.line)
```
